```python
from bs4 import BeautifulSoup
import requests
import time
from datetime import date, timedelta
import shutil, sys 
```

```python
def download(name, file):
    image_url = name
    filename="c:/data/tapas/"+file+".jpg"
    # Open the url image, set stream to True, this will return the stream content.
    r = requests.get(image_url, stream = True)

    # Check if the image was retrieved successfully
    if r.status_code == 200:
        # Set decode_content value to True, otherwise the downloaded image file's size will be zero.
        r.raw.decode_content = True
    
        # Open a local file with wb ( write binary ) permission.
        with open(filename,'wb') as f:
            shutil.copyfileobj(r.raw, f)
        
        print('Image sucessfully Downloaded: ',file)
    else:
        print('Image Couldn\'t be retreived')
        print(name)
        
def daterange(start_date, end_date):
    for n in range(int((end_date - start_date).days)):
        yield start_date + timedelta(n)
```

```python
start_date = date(1945, 8, 28)
end_date = date(2021, 7, 4)
#end_date = date(1945, 9, 4)
for sd in daterange(start_date, end_date):    
    file=sd.strftime("%Y")+sd.strftime("%m")+sd.strftime("%d")
    name="https://tapas.clarin.com/tapa"+"/"+sd.strftime("%Y")+"/"+sd.strftime("%m")+"/"+sd.strftime("%d")+"/"+file+".jpg"
    download(name, file)
```
