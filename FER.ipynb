```python
from fer import FER
import cv2
import json
import os
```

```python
import numpy as np 
class npEncoder(json.JSONEncoder):
    def default(self, obj):
        if isinstance(obj, np.intc):
            return int(obj)
        if isinstance(obj, np.int32):
            return int(obj)
        return json.JSONEncoder.default(self, obj)
```

```python
db=[]
detector = FER()
for archivo in os.listdir("C:/data/tapas/"):
  if archivo.endswith('.jpg'):
    img = cv2.imread("C:/data/tapas/" + str(archivo))
    try:
        result=detector.detect_emotions(img)
        print(archivo)
        new={}
        new["archivo"]=archivo
        for a in result: a.update(new)
        db+=result
    except:
        print("no se pudo obtener " + archivo)

with open('rostros.json', 'w') as fout:
    json.dump(db, fout, cls=npEncoder)
```
