```python
import bar_chart_race as bcr
import matplotlib.font_manager
import pandas as pd
import datetime
```

```python
df = pd.read_csv("C:/Users/PC/Clarin/Clarin.csv", encoding="latin-1", sep=";")
```

```python
df['year'] = df['file'].astype(str).str[0:4]
df['month'] = df['file'].astype(str).str[4:6]
df['day'] = df['file'].astype(str).str[6:8]
df['fecha'] = pd.to_datetime(df['month']+"/"+df['day']+"/"+df['year'], format='%m/%d/%Y')
```

```python
df.drop(df.columns.difference(['text','fecha']), 1, inplace=True)
```

```python
df
```

```python
yearly=df.resample('Y', on='fecha').sum()
```

```python
bcr.bar_chart_race(
    df=df,
    filename='prueba.mp4',
    orientation='h',
    sort='desc',
    n_bars=7,
    fixed_order=False,
    fixed_max=True,
    steps_per_period=10,
    interpolate_period=False,
    label_bars=True,
    bar_size=.95,
    period_label={'x': .99, 'y': .25, 'ha': 'right', 'va': 'center'},
    period_fmt='%B %d, %Y',
    period_summary_func=lambda v, r: {'x': .99, 'y': .18,
                                      's': f'Total deaths: {v.nlargest(6).sum():,.0f}',
                                      'ha': 'right', 'size': 8, 'family': 'Courier New'},
    perpendicular_bar_func='median',
    period_length=500,
    figsize=(5, 3),
    dpi=144,
    cmap='dark12',
    title='COVID-19 Muertes por país',
    title_size='',
    bar_label_size=7,
    tick_label_size=7,
    shared_fontdict={'family' : 'Courier New', 'color' : '.1'},
    scale='linear',
    writer=None,
    fig=None,
    bar_kwargs={'alpha': .7},
    filter_column_colors=False)  
```
